Recurrent Neural Network

Recurrent neural networks are specifically designed to work with sequence data.

So examples of sequence data are things like time series data, such as how many sales have happened
over a previous time series or what the production has been over a previous time series.

There's also things like sentences, which is what we're focusing on for natural language processing. So you can think of a sentence as just a sequence of words.

Things like audio or music. It's basically just a sequence of sounds or even things like car trajectories.
It's just a sequence of instructions; Left, right, forward, and back.

Cells that are a function of inputs from previous time steps are also known as **memory cells**

RNN are also flexible in their inputs and outputs.

Eg: 
- Sequence to Sequence:  we can actually perform a sequence input to a sequence output. And an example of that would be passing in a set of time series information, such as a year's worth of daily sales data. And then wanting back a sequence of that same sales data shifted over a certain time period into the future.

- Sequence to vector: set of examples, inputs and outputs, would be to pass in a sequence input and request back a vector output. a common example of using recurring neural networks for this sort of input and output is sentiment scores. So you can feed in a sequence of words, maybe a paragraph of a movie review, and then request back a vector, indicating whether it was a positive sentiment such as, they really like the movie.So that's usually indicated by one versus they really dislike the movie, which is indicated by usually negative one or zero.

    So on the opposite side, you could also feed in a vector input. So a single input at just a first time step
        and basically pass in zeros for the rest of your time steps. And then let the output be a sequence. **vector to sequence network**... So an example of vector to sequence could be just providing a single seed word and then getting out an entire sequence of high-probability sequence phrases that would actually come out. So you could have a word such as "Hello" and then the sequence that gets generated something like, "How are you?" Because after being trained on lots of introductory text, maybe the neural network realizes that just based off the seed, "Hello," one of the more common phrases is, "How are you?"

LSTMs, GRU, and Text Generation

# LSTM - Long Short Term Memory Units

are going to be necessary in order to actually generate text that makes sense because we wanna have the network be aware of not just the most recent text it's seen, but also the entire history of text that it's seen.



So a common issue that recurrent neural networks face is that after a while, especially if you're training the network on a really large sequence, then the recurrent neural network is going to begin to forget those very first inputs. The later and later training batches of later inputs, things that come towards the end of the text document, that's gonna start kind of overriding the weights of the very beginning inputs
of the beginning of the text. So what we wanna make sure is that we're not beginning to forget those first inputs as information is lost at each step going through the recurrent neural network. So what we really need is we need some sort of long term memory for our networks. We need to balance both the short term memory of the networks, the data that it was recently trained on, versus the long term memory of the networks, all the data and the very first data that was being trained on. So the LSTM, or long short-term memory cell, was created to help address these recurrent neural network issues.

Variants on an LSTM cell

- There's a variant called the peep hole variance. this is actually quite popular, and the reason for that is because it adds peepholes to all the gates.

- **GRU** - Gated Recurrent Unit.  What this ends up doing is it actually simplifies things a bit by combining the forget and input gates into a single, what they call, update gates. And it also merges the cell state and hidden state, and it makes a few other changes. So this resulting model is actually simpler than standard LSTM models.


Depth Gated Recurrent Neural Networks, and that was introduced in 2015.

for text generation LSTMs work best.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

```python
vocabulary_size
```

```python
def create_model(vocabulary_size, seq_len):

    model = Sequential()
    # Turns positive integers (indexes) into dense vectors of fixed size.
    model.add(Embedding(vocabulary_size,
                        seq_len,
                        input_length=seq_len))
    # LSTM units to actually be trained on the sequence.
    model.add(LSTM(units=seq_len*2,
                   return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(50, activation="relu"))
    
    # translate this into a vocabulary word because it's just going to spit out a number.
    model.add(Dense(vocabulary_size, activation="softmax"))

    model.compile(loss="categorical_crossentropy",
                  optimizer="adam",
                  metrics=['accuracy'])
    
    model.summary()

    return model
```

# Generating New Text

```python
from random import randint
from pickle import load
from keras.models import load_model
from keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    '''
    INPUTS:
    model : model that was trained on text data
    tokenizer : tokenizer that was fit on text data
    seq_len : length of training sequence
    seed_text : raw string text to serve as the seed
    num_gen_words : number of words to be generated by model
    '''
    
    # Final Output
    output_text = []
    
    # Intial Seed Sequence
    input_text = seed_text
    
    # Create num_gen_words
    for i in range(num_gen_words):
        
        # Take the input text string and encode it to a sequence
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        # Pad sequences to our trained rate
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        
        # Predict Class Probabilities for each word
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        
        # Grab word
        pred_word = tokenizer.index_word[pred_word_ind] 
        
        # Update the sequence of input text (shifting one over with the new word)
        input_text += ' ' + pred_word
        
        output_text.append(pred_word)
        
    # Make it look like a sentence.
    return ' '.join(output_text)
```
